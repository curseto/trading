```python
import sys
from PyQt5.QtWidgets import *

app = QApplication(sys.argv) # QApplication 클래스의 인스턴스가 GUI 프로그램을 전반적으로 제어하는 역할
label = QLabel("Hello PyQt") # UI 구성을 담당, QLabel클래스의 인스턴스 생성
label.show()                 # QLabel클래스의 show 매서드 호출
app.exec_()
```

## QMainWindow를 이용한 윈도우 생성

```python
import sys
from PyQt5.QtWidgets import *

class MyWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("PyStock")
        self.setGeometry(300, 300, 300, 400)

if __name__ == "__main__":
    app = QApplication(sys.argv)
    mywindow = MyWindow()
    mywindow.show()
    app.exec_()
```

## 이벤트처리

```python
import sys
from PyQt5.QtWidgets import *
from PyQt5.QtCore import *

class MyWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("PyStock")        # mywindow 객체를 통해 매서드 사용(매서드는 QMainWindw클래스에 존재)
        self.setGeometry(300, 300, 300, 400)  # mywindow 객체를 통해 매서드 사용(매서드는 QMainWindw클래스에 존재)

        btn1 = QPushButton("Click me", self)  # btn1객체 생성, (self는 버튼이 MyWindow 객체(즉, 부모 위젯) 위에 생성되도록 설정합니다.)
        btn1.move(20, 20)                     # btn1객체를 통해 move매서드 사용
        btn1.clicked.connect(self.btn1_clicked)

    def btn1_clicked(self):
        QMessageBox.about(self, "message", "clicked") # QMessageBox는 클래스 about은 이 클래스의 매서드

if __name__ == "__main__":
    app = QApplication(sys.argv)
    myWindow = MyWindow()    # MyWindow클래스의 myWindow객체 생성
    myWindow.show()
    app.exec_()
```

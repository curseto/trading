## API

- API(Application Programming Interface)란 파이썬의 함수나 메서드와 같은 것입니다.
- 증권사 API는 파이썬으로 구현되어있지 않다.
- 다른 프로그래밍 언어로 작성된 함수를 파이썬에서 어떻게 호출할 수 있는가?
- COM과 OCX

### COM(Component Object Model)
- COM이란 마이크로소프트에서 개발한 기술입니다
- COM은 여기서 더 나아가 프로그래밍 언어와 상관없이 개발된 객체를 사용할 수 있게 해줍니다
- Dispatch 함수가 호출되면 객체가 생성되는데 excel이라는 변수가 생성된 COM 객체를 바인딩합니다.

```python
import win32com.client
excel = win32com.client.Dispatch("Excel.Application")


excel.Visible = True
```

- 파이썬에서 다른 프로그래밍 언어로 작성된 COM 객체를 생성하려면 win32com.client라는 모듈의 Dispatch 메서드를 사용하면 됩니다
- Dispatch 함수에 의해 Excel.Application라는 COM 객체가 생성되고 excel이라는 변수가 이 COM 객체를 바인딩함

```python
import win32com
print(win32com.__file__) # win32com 패키지 폴더의 디렉토리 반환
```

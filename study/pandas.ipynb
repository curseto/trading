#### replace()
- The replace method is used to replace values in a DataFrame or Series
- Setting regex=True treats the pattern as a regular expression.

```python
import pandas as pd

# Example DataFrame
data = {
    "a": ["--10", "20", "--30"],
    "b": ["--5", "15", "--25"],
    "c": ["--8", "--18", "28"],
    "d": ["0", "--1", "--2"],
    "e": ["Unchanged", "Value", "--50"]  # Column not to be modified
}
df = pd.DataFrame(data)

# Replace "--" with "-" in specific columns
columns_to_replace = ["a", "b", "c", "d"]
df[columns_to_replace] = df[columns_to_replace].replace("--", "-", regex=True)

print(df)
```

#### concat()
- If you have multiple DataFrames stored in a list, you can combine them into a single DataFrame using pd.concat.
- By default, pd.concat combines DataFrames vertically (row-wise).

```python
import pandas as pd

# Create two DataFrames
df1 = pd.DataFrame({"A": [1, 2], "B": [3, 4]})
df2 = pd.DataFrame({"A": [5, 6], "B": [7, 8]})

# Create an empty list
dataframes = []

# Append DataFrames to the list
dataframes.append(df1)
dataframes.append(df2)

print(dataframes)
print('-------------')
# Combine DataFrames in the list into a single DataFrame
df_combined = pd.concat(dataframes)
print(df_combined)
```

## Statistics

#### rolling
- method provided by Pandas for calculating rolling (or moving) statistics, such as moving averages, over a specified window of time

DataFrame.rolling(window=N).<aggregation_function>()
- window=N: Specifies the size of the rolling window (e.g., 10 for a 10-day rolling average)
- \<aggregation_function\>: The function applied to the rolling window (e.g., .mean(), .sum())

```python
import pandas as pd

# Example DataFrame
data = {"Close": [30500, 30700, 26500, 27850, 26000, 25750, 24300, 25950, 24500, 20800]}
df = pd.DataFrame(data)

# Compute 10-day moving average
df["10_day_MA"] = df["Close"].rolling(window=5).sum()
print(df)
```
